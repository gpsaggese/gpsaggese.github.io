# Week 1 – Exploratory Data Analysis (EDA)
**Goal:** Explore the Drug-Drug Interaction dataset and understand its structure before preprocessing.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
df = pd.read_csv('../data/sample.csv')
df.head()
```

```python
# Basic info
print("Shape:", df.shape)
df.info()
df.describe(include='all')
```

## 🔍 Observations
- Each row represents an interaction between two drugs.
- Columns:
  - **drugA**, **drugB**: Drug names
  - **interaction_type**: Category of their interaction


```python
# Visualize distribution of interaction types
sns.countplot(data=df, x='interaction_type', palette='cool')
plt.title('Distribution of Interaction Types')
plt.show()
```

## ✅ Summary
- Dataset loaded successfully.
- Interaction types show clear categories.
