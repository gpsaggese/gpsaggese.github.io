```python

from sentiment_utils import prepare_data, vectorize, train_model, evaluate
import pandas as pd
```

```python
class SentimentAPI:
    """High-level wrapper for the sentiment analysis workflow."""
    def __init__(self):
        self.vectorizer = None
        self.model = None

    def fit(self, df, text_col="review", label_col="sentiment"):
        X_train, X_test, y_train, y_test = prepare_data(df, text_col, label_col)
        X_train_vec, X_test_vec, self.vectorizer = vectorize(X_train, X_test)
        self.model = train_model(X_train_vec, y_train)
        return evaluate(self.model, X_test_vec, y_test)

    def predict(self, text):
        vec = self.vectorizer.transform([text])
        return self.model.predict(vec)[0]
```

```python
data = pd.DataFrame({
    "review": ["Amazing phone!", "Horrible battery life", "Average camera quality"],
    "sentiment": ["positive", "negative", "neutral"]
})
api = SentimentAPI()
report = api.fit(data)
print(report)
```
