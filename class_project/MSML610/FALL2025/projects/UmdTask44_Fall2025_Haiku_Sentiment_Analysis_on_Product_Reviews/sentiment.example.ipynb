```python
import pandas as pd

# Load dataset
df = pd.read_csv("data/Reviews.csv")

# Display shape and first few rows
print(df.shape)
df.head()
```

# Checking Missing Values

```python
df.isnull().sum()
```

```python
df["Text_Length"] = df["Text"].apply(len)
df["Text_Length"].describe()
```

# Rating Distribution

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
df["Score"].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title("Distribution of Review Scores")
plt.xlabel("Score (1 = Negative, 5 = Positive)")
plt.ylabel("Count")
plt.show()
```

# Relationship between Score and Text Length

```python
df.groupby("Score")["Text_Length"].mean().plot(kind='bar', color='orange')
plt.title("Average Review Length by Rating")
plt.show()
```

```python
# Word Cloud
```

```python
from wordcloud import WordCloud

text = " ".join(review for review in df["Text"].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
