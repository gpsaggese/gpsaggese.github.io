```python
# Example use of the `SentimentAPI` to train and evaluate a sentiment classifier on a sample dataset.

import pandas as pd
from sentiment_utils import prepare_data, vectorize, train_model, evaluate

df = pd.read_csv("https://raw.githubusercontent.com/dD2405/Twitter_Sentiment_Analysis/master/train.csv")
df = df.rename(columns={"label": "sentiment", "tweet": "review"})
df.head()
```

```python
X_train, X_test, y_train, y_test = prepare_data(df, "review", "sentiment")
X_train_vec, X_test_vec, vec = vectorize(X_train, X_test)
model = train_model(X_train_vec, y_train)
report = evaluate(model, X_test_vec, y_test)
print(report)
```

```python
sample = "This product exceeded my expectations!"
pred = model.predict(vec.transform([sample]))[0]
print("Predicted sentiment:", pred)
```
