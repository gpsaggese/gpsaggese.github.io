# Electricity Load Forecasting - Complete Project

Full end-to-end forecasting with AutoKeras.

```python
import sys
sys.path.append('../utils')

from autokeras_utils import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print('✅ Setup complete')
```

## Step 1: Load Data

```python
preprocessor = ElectricityDataPreprocessor()
df = preprocessor.load_and_prepare_data('../data/PJME_hourly.csv', 'Datetime', 'PJME_MW')

print(f'Dataset: {df.shape}')
print(f'Period: {df.index.min()} to {df.index.max()}')
df.head()
```

## Step 2: Visualize Data

```python
plt.figure(figsize=(15, 6))
plt.plot(df.index, df['PJME_MW'])
plt.title('PJM Electricity Load')
plt.xlabel('Date')
plt.ylabel('Load (MW)')
plt.show()

print(df['PJME_MW'].describe())
```

## Step 3: Feature Engineering

```python
print('Engineering features...')
df_features = preprocessor.prepare_features(df, 'PJME_MW', create_lags=True, create_rolling=True)

print(f'✅ Total features: {df_features.shape[1]}')
print(f'Features: {df_features.columns.tolist()[:10]}...')
```

## Step 4: Split and Scale Data

```python
X_train, X_val, X_test, y_train, y_val, y_test = preprocessor.split_data(df_features, 'PJME_MW', test_size=0.2, val_size=0.1)
X_train_s, X_val_s, X_test_s = preprocessor.normalize_features(X_train, X_val, X_test)

print(f'Train: {len(X_train)} | Val: {len(X_val)} | Test: {len(X_test)}')
```

## Step 5: Train AutoKeras (20-30 minutes)

```python
forecaster = AutoKerasForecaster(max_trials=10, epochs=100, seed=42)
forecaster.build_model(X_train_s, y_train)

print('🚀 Training AutoKeras (this takes 20-30 minutes)...')
forecaster.train(X_train_s, y_train, X_val_s, y_val, verbose=1)
print('✅ Training complete!')
```

## Step 6: Baseline Models

```python
naive_pred = BaselineModels.naive_forecast(y_train, len(y_test))
seasonal_pred = BaselineModels.seasonal_naive_forecast(y_train, len(y_test), season_length=24)
ma_pred = BaselineModels.moving_average_forecast(y_train, len(y_test), window=24)

print('✅ Baselines computed')
```

## Step 7: Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

def evaluate(name, pred):
    mae = mean_absolute_error(y_test, pred)
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    mape = mean_absolute_percentage_error(y_test, pred) * 100
    print_evaluation_metrics({'MAE': mae, 'RMSE': rmse, 'MAPE': mape}, name)
    return {'MAE': mae, 'RMSE': rmse, 'MAPE': mape}

ak_pred = forecaster.predict(X_test_s)
ak_metrics = evaluate('AutoKeras', ak_pred)
naive_metrics = evaluate('Naive', naive_pred)
seasonal_metrics = evaluate('Seasonal', seasonal_pred)
ma_metrics = evaluate('Moving Avg', ma_pred)
```

## Step 8: Visualizations

```python
viz = ForecastVisualizer()

viz.plot_predictions(y_test, ak_pred, 'AutoKeras Forecast')
plt.show()
```

```python
viz.plot_error_distribution(y_test, ak_pred)
plt.show()
```

```python
metrics_dict = {
    'AutoKeras': ak_metrics,
    'Naive': naive_metrics,
    'Seasonal': seasonal_metrics,
    'Moving Avg': ma_metrics
}

viz.plot_metrics_comparison(metrics_dict)
plt.show()
```
