# Weights & Biases (W&B) API Demo Notebook
- This notebook serves as a tutorial for the `wandb` Python SDK. It demonstrates the core API functions—authentication, initialization, configuration logging, and metric logging—using a simple, project-agnostic example.

```python
import wandb
import random
import time

print(f"W&B SDK Version: {wandb.__version__}")
```

## 1. Authentication (`wandb.login()`)

W&B requires authentication using an API key. It is best practice to set the API key as the `WANDB_API_KEY` environment variable.

Run the cell below to authenticate your environment.

```python
# This command authenticates your machine using the WANDB_API_KEY environment variable.
wandb.login()
```

## 2. Initialization and Configuration (`wandb.init()` and `wandb.config`)

The `wandb.init()` function starts a new **Run**. The `config` parameter logs the **static hyperparameters** that define the experiment (e.g., model type, batch size).

```python
PROJECT_NAME = "API-Demo-Project"

# Define a generic set of hyperparameters (config) to be logged once
generic_config = {
    "model_type": "SimpleFFN",
    "learning_rate": 0.001,
    "batch_size": 32,
    "epochs": 10,
    "optimizer": "Adam"
}

# Start a new W&B Run and log the configuration
run = wandb.init(
    project=PROJECT_NAME,
    name="simulated_run_A",
    config=generic_config
)

print(f"Run URL: {run.url}")

```

## 3. Logging Metrics
- Use `wandb.log()` to stream dynamic metrics (loss, accuracy, etc.) inside a loop. Each call creates a new step. `run.finish()` closes the run.

```python
NUM_STEPS = run.config.epochs * 2  # Log metrics 2 times per epoch

print("\nStarting simulated process...")

for step in range(1, NUM_STEPS + 1):
    # Simulate calculation of dynamic metrics
    current_loss = 1.0 - (step * 0.04) + random.uniform(-0.01, 0.01)
    current_accuracy = (step * 0.04) + random.uniform(-0.01, 0.01)
    
    # Log the metrics to W&B
    wandb.log({
        "loss": current_loss,
        "accuracy": current_accuracy,
        "process_step": step
    })
    
    # Print status to console for local tracking
    if step % 4 == 0:
        print(f"Step: {step}, Loss: {current_loss:.4f}, Accuracy: {current_accuracy:.4f}")
        
    time.sleep(0.1)
    
print("\nSimulated process complete.")

# Ensure the run is finalized and all data is uploaded
run.finish()
```

## Summary

This notebook successfully demonstrated the four core steps of the W&B API for experiment tracking:

1.  **`wandb.login()`**: Authentication.
2.  **`wandb.init(project=..., config=...)`**: Starting a Run and logging static configuration.
3.  **`wandb.log(...)`**: Streaming dynamic metrics in a simulated loop.
4.  **`run.finish()`**: Closing the run.
