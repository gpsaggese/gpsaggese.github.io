```python
import requests

API_URL = "http://172.16.7.95:8000"  # change if running remotely
API_URL = "http://localhost:8000"

def get_image_embedding(image_path: str):
    """
    Sends an image to the /embed/image endpoint and returns its embedding.
    """
    url = f"{API_URL}/embed/image"
    with open(image_path, "rb") as img_file:
        files = {"image": (image_path, img_file, "image/jpeg")}
        response = requests.post(url, files=files)

    if response.status_code == 200:
        data = response.json()
        # print(f"Image embedding received for: {data['image_filename']}")
        return data["embedding"]
    else:
        print("Error:", response.text)
        return None


def get_text_embedding(text: str):
    """
    Sends text to the /embed/text endpoint and returns its embedding.
    """
    url = f"{API_URL}/embed/text"
    data = {"text": text}
    response = requests.post(url, data=data)

    if response.status_code == 200:
        data = response.json()
        # print(f"Text embedding received for: {data['text']}")
        return data["embedding"]
    else:
        print("Error:", response.text)
        return None


# if __name__ == "__main__":
#     # Example usage
#     image_path = "cat.jpg"
#     text_input = "A cute cat sitting on the floor"

#     image_embed = get_image_embedding(image_path)
#     text_embed = get_text_embedding(text_input)

#     print("\nImage embedding length:", len(image_embed))
#     print("Text embedding length:", len(text_embed))
```

### Figures Indexing

```python
%%time
for pth in tqdm(dataset['image'].unique()):
    img_embed = get_text_embedding(f"flickr8k/Images/{pth}")
    norm_ = np.linalg.norm(img_embed)
    if norm_ > 0:
        img_embed = img_embed / norm_
    else:
        img_embed = img_embed

    img_embed = np.expand_dims(img_embed, axis=0)
    image_index.add(img_embed)
```

```python
%%time
embedding_dim = 512
n_images = len(dataset['image'].unique())

image_embs_mmap = np.memmap('image_embs.dat', dtype='float32', mode='w+', shape=(n_images, embedding_dim))

image_index = faiss.IndexFlatIP(embedding_dim)

for i, pth in enumerate(tqdm(dataset['image'].unique())):
    # Get embedding
    img_embed = get_text_embedding(f"flickr8k/Images/{pth}")

    norm_ = np.linalg.norm(img_embed)
    if norm_ > 0:
        img_embed = img_embed / norm_

    img_embed_faiss = np.expand_dims(img_embed.astype('float32'), axis=0)
    image_index.add(img_embed_faiss)

    image_embs_mmap[i] = img_embed.astype('float32')

# Flush memmap to disk
image_embs_mmap.flush()

# Save FAISS index
faiss.write_index(image_index, "image_index.faiss")
```

### Text Indexing

```python
%%time
for txt in tqdm(dataset['caption'].unique()):
    txt_embed = get_text_embedding(txt)
    norm_ = np.linalg.norm(txt_embed)
    if norm_ > 0:
        txt_embed = txt_embed / norm_
    else:
        txt_embed = txt_embed

    txt_embed = np.expand_dims(txt_embed, axis=0)
    text_index.add(txt_embed)
```

```python
df = dataset.copy()
```

```python
caption_to_image = dict(enumerate(df['image']))

# Build mapping from image to all its captions
from collections import defaultdict
image_to_captions = defaultdict(list)
for idx, row in df.iterrows():
    image_to_captions[row['image']].append(idx)
```

```python
import numpy as np
import faiss

# Load your FAISS index
image_index = faiss.read_index("image_index.faiss")

# Suppose you stored the image order separately
image_list = df['image'].unique().tolist()
image_id_to_idx = {img: i for i, img in enumerate(image_list)}

k = 5
recalls = []
precisions = []

i = 0
for cap_idx, row in tqdm(df.iterrows()):
    query_txt = get_text_embedding(row['caption'])
    query_txt = query_txt / np.linalg.norm(query_txt)
    query_txt = np.expand_dims(query_txt, axis=0)

    distances, indices = image_index.search(query_txt, k)

    retrieved_imgs = [image_list[i] for i in indices[0]]
    true_img = row['image']

    # Compute precision/recall@K for this query
    relevant = {true_img}
    retrieved = set(retrieved_imgs)

    tp = len(relevant & retrieved)
    precision = tp / k
    recall = tp / len(relevant)  # always 1 relevant item per caption

    precisions.append(precision)
    recalls.append(recall)
    i+=1
    if i == 100:
        break

print(f"Mean Precision@{k}: {np.mean(precisions):.3f}")
print(f"Mean Recall@{k}: {np.mean(recalls):.3f}")
```
