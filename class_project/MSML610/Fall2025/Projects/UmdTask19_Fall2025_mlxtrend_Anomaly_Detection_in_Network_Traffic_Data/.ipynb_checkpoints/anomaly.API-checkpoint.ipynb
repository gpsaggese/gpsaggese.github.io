```python
from anomaly_utils import load_unsw_from_zip, basic_eda

df = load_unsw_from_zip("archive.zip", extract_dir="./data")
basic_eda(df)
df["label"].value_counts()
```

## Building our preprocessing and feature-selection API layer

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

y = df["label"].astype(int)
X = df.drop(columns=["label"])

cat_cols = ["proto", "service", "state", "attack_cat"]
num_cols = [c for c in X.columns if c not in cat_cols]

numeric_tf = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler(with_mean=False))
])

categorical_tf = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=True))
])

preprocess = ColumnTransformer([
    ("num", numeric_tf, num_cols),
    ("cat", categorical_tf, cat_cols)
])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)

print("Preprocessing pipeline ready.")
print("Train:", X_train.shape, " Test:", X_test.shape)
```

```python
!pip install mlxtend --no-cache-dir
```

```python
from sklearn.feature_selection import mutual_info_classif
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import pandas as pd, numpy as np, time, json, os

numeric_mask = X_train.apply(lambda s: np.issubdtype(s.dtype, np.number))
num_cols_valid = X_train.columns[numeric_mask].tolist()

mi = mutual_info_classif(X_train[num_cols_valid].fillna(0), y_train, random_state=42)
mi_series = pd.Series(mi, index=num_cols_valid).sort_values(ascending=False)
top_num = mi_series.head(10).index.tolist()

print("Top numeric features by Mutual Information:")
print(top_num)

imp = SimpleImputer(strategy="median")
scal = StandardScaler(with_mean=False)
Xs = scal.fit_transform(imp.fit_transform(X_train[top_num]))

clf = LogisticRegression(max_iter=500, solver="liblinear")
efs = EFS(clf, min_features=4, max_features=6, scoring="roc_auc", cv=3, n_jobs=1)
t0 = time.time()
efs.fit(Xs, y_train)
print(f"EFS took {(time.time()-t0)/60:.1f} min.")

selected_numeric = [top_num[i] for i in efs.best_idx_]
print("✅ Selected numeric features:", selected_numeric)

os.makedirs("outputs", exist_ok=True)
json.dump(selected_numeric, open("outputs/selected_numeric.json", "w"))
```
