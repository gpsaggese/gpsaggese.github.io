```python
from anomaly_utils import load_unsw_from_zip
from sklearn.model_selection import train_test_split
import pandas as pd, numpy as np, json

# === Step 1: Load dataset ===
df = load_unsw_from_zip("archive.zip", extract_dir="./data")

# Take a random sample for modeling (prevents kernel crashes)
df = df.sample(10000, random_state=42).reset_index(drop=True)
print(f"⚡ Using subset of {len(df):,} rows for modeling.")

# === Step 2: Define target and features ===
y = df["label"].astype(int)
X = df.drop(columns=["label"])

# === Step 3: Separate numeric and categorical columns ===
cat_cols = ["proto", "service", "state", "attack_cat"]
num_cols = [c for c in X.columns if c not in cat_cols]

# === Step 4: Load previously selected numeric features ===
selected_numeric = json.load(open("outputs/selected_numeric.json"))
print("Loaded selected numeric features:", selected_numeric)

# Display sample info
print("\n📊 Dataset summary:")
print(f"Features: {len(X.columns)} | Numeric: {len(num_cols)} | Categorical: {len(cat_cols)}")
print("Label distribution:")
print(y.value_counts(normalize=True))
```
