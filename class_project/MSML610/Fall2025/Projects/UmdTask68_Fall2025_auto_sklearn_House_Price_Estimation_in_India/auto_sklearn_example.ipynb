```python
# setup and imports
import sys
import pandas as pd
import numpy as np
from auto_sklearn_utils import prepare_data
```

```python
# define data path
data_path = "data/raw/india_housing_prices.csv"
print(f"loading data from: {data_path}")
```

```python
# prepare data
X_train_processed, X_test_processed, y_train, y_test, preprocessor, feature_names = prepare_data(
    data_path=data_path,
    target_col="Price_in_Lakhs",
    test_size=0.2,
    random_state=42
)

print("data preparation complete!")
```

```python
# display dataset shapes
print(f"training Features Shape:   {X_train_processed.shape}")
print(f"test Features Shape:       {X_test_processed.shape}")
print(f"training Target Shape:     {y_train.shape}")
print(f"test Target Shape:         {y_test.shape}")
print(f"\ntotal Features Engineered: {len(feature_names)}")
```

```python
# display feature names
print(f"total features: {len(feature_names)}\n")
print("first 20 features:")
for i, fname in enumerate(feature_names[:20], 1):
    print(f"  {i:2d}. {fname}")

print("\n...")
print(f"\nlast 10 features:")
for i, fname in enumerate(feature_names[-10:], len(feature_names) - 9):
    print(f"  {i:2d}. {fname}")
```

```python
# display target variables 
print("training target (Price in Lakhs) statistics:")
print(y_train.describe())

print("\ntest target (Price in Lakhs) statistics:")
print(y_test.describe())
```

```python
# display processed data samples
print("first 5 rows of processed training features:")
print(X_train_processed[:5])

print("\ndata type information:")
print(f"X_train data type: {X_train_processed.dtype}")
print(f"X_test data type: {X_test_processed.dtype}")

print("\nmissing values:")
print(f"NaN in X_train: {np.isnan(X_train_processed).sum()}")
print(f"NaN in X_test: {np.isnan(X_test_processed).sum()}")
```

```python
# display basic stats of processed features
# convert to DataFrame for easy analysis
X_train_df = pd.DataFrame(X_train_processed, columns=feature_names)

print("basic statistics of processed training features:")
print(X_train_df.describe())

print(f"\nfeature value ranges (min, max):")
print(f"min value across all features: {X_train_processed.min():.4f}")
print(f"max value across all features: {X_train_processed.max():.4f}")
print(f"mean value across all features: {X_train_processed.mean():.4f}")
print(f"std dev across all features: {X_train_processed.std():.4f}")
```
