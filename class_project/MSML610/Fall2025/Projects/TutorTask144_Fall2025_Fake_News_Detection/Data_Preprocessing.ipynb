# Data Preprocessing for Fake News Detection

This notebook demonstrates the complete data preprocessing pipeline including:
1. Loading raw data from data/true.csv and data/fake.csv
2. Text normalization (lowercase, remove URLs, special characters, **remove numbers**)
3. Tokenization (convert text to words)
4. Stopword removal (remove common words)
5. Stemming (reduce words to root form)
6. Analysis and statistics

## Part 1: Setup and Import

```python
import preprocessing
import bert_utils
import pandas as pd
import numpy as np
import warnings
from pathlib import Path
import time

warnings.filterwarnings('ignore')
```

## Part 2: Load Raw Data


```python
df = bert_utils.load_base_dataset(data_dir='data')
```

```python
print("SAMPLE RAW TEXT (Real Article)")

sample_real = df[df['label'] == 1]['content'].iloc[0]
print(sample_real[:500])
print(f"\nLength: {len(sample_real)} characters")
```

```python
print("SAMPLE RAW TEXT (Fake Article)")

sample_fake = df[df['label'] == 0]['content'].iloc[0]
print(sample_fake[:500])
print(f"\nLength: {len(sample_fake)} characters")
```

## Part 3: Demonstrate Individual Preprocessing Steps

```python
print("PREPROCESSING PIPELINE DEMONSTRATION\n")

print(f"\nOriginal text:")
print(f"  {sample_real}")
print(f"  Length: {len(sample_real)} characters")
```

### Step 1: Normalization

```python
normalized = preprocessing.normalize_text(sample_real)
print(f"  {normalized}")
print(f"  Length: {len(normalized)} characters")
```

### Step 2: Tokenization

```python
tokens = preprocessing.tokenize(normalized)
print(f"  Tokens: {tokens}")
print(f"  Number of tokens: {len(tokens)}")
```

### Step 3: Stopword Removal


```python
filtered = preprocessing.remove_stopwords(tokens)
print(f"  Tokens: {filtered}")
print(f"  Number of tokens: {len(filtered)}")
print(f"  Removed: {len(tokens) - len(filtered)} stopwords")
```

### Step 4: Stemming (Porter Stemmer)


```python
stemmed = preprocessing.stem_tokens(filtered)
print(f"\n[Step 4] After Stemming:")
print(f"  Tokens: {stemmed}")
print(f"  Number of tokens: {len(stemmed)}")
print(f"\n  Example stem transformations:")
for orig, stem in zip(filtered, stemmed):
    if orig != stem:
        print(f"    {orig} → {stem}")
```

## Part 4: Apply Full Pipeline to All Data

```python
df_preprocessed = df.copy()
df_preprocessed = preprocessing.preprocess_dataframe(
    df,
    text_column='content',
    normalize=True,
    tokenize_text=True,
    remove_stops=True,
    stem=True
)

print(f"  Articles processed: {len(df_preprocessed):,}")
```

## Part 5: Preprocessing Statistics

```python
stats = preprocessing.get_preprocessing_stats(df, df_preprocessed, text_column='content')

print("PREPROCESSING STATISTICS - FULL DATASET (44,898 articles)\n\n")

print(f"Original samples:           {stats['original_samples']:,}")
print(f"Preprocessed samples:       {stats['preprocessed_samples']:,}")
print(f"Samples removed:            {stats['samples_removed']}")

print(f"\nOriginal text statistics:")
print(f"  Average length:           {stats['avg_original_length']:.0f} characters")
print(f"  Min:                      {df['content'].apply(len).min():,} characters")
print(f"  Max:                      {df['content'].apply(len).max():,} characters")

print(f"\nPreprocessed text statistics:")
print(f"  Average length:           {stats['avg_preprocessed_length']:.0f} characters")
print(f"  Min:                      {df_preprocessed['content'].apply(len).min():,} characters")
print(f"  Max:                      {df_preprocessed['content'].apply(len).max():,} characters")

print(f"\nData reduction:")
print(f"  Text length reduction:    {stats['length_reduction_pct']:.1f}%")

print(f"\nLabel distribution (after preprocessing):")
real_count = (df_preprocessed['label'] == 1).sum()
fake_count = (df_preprocessed['label'] == 0).sum()
print(f"  Real articles (label=1):  {real_count:,}")
print(f"  Fake articles (label=0):  {fake_count:,}")
```

## Part 6: Before & After Comparison

```python
for label_val, label_name in [(1, "REAL"), (0, "FAKE")]:
    subset = df[df['label'] == label_val]
    sample_idx = subset.sample(n=1, random_state=42).index[0]
    
    original = df.loc[sample_idx, 'content']
    preprocessed = df_preprocessed.loc[sample_idx, 'content']
    
    print(f"\n[{label_name} Article - Label={label_val}]")
    print(f"\nBEFORE (Original):")
    print(f"  {original[:150]}...")
    print(f"  Length: {len(original)} chars | Words: {len(original.split())}")
    print(f"\nAFTER (Preprocessed):")
    print(f"  {preprocessed[:150]}...")
    print(f"  Length: {len(preprocessed)} chars | Words: {len(preprocessed.split())}")
    print(f"  Reduction: {100 * (1 - len(preprocessed) / len(original)):.1f}%")
```

## Part 7: Save Preprocessed Data

```python
import os
output_path = 'data/preprocessed_data.csv'
df_preprocessed.to_csv(output_path, index=False)

file_size_mb = os.path.getsize(output_path) / (1024*1024)
file_size_kb = os.path.getsize(output_path) / 1024

print(f"\nOutput file: {output_path}")
print(f"File size: {file_size_mb:.2f} MB ({file_size_kb:.1f} KB)")
print(f"Total articles: {len(df_preprocessed):,}")
print(f"  - Real (label=1): {(df_preprocessed['label'] == 1).sum():,}")
print(f"  - Fake (label=0): {(df_preprocessed['label'] == 0).sum():,}")
print(f"\nDataFrame shape: {df_preprocessed.shape}")
print(f"Columns: {list(df_preprocessed.columns)}")
```
