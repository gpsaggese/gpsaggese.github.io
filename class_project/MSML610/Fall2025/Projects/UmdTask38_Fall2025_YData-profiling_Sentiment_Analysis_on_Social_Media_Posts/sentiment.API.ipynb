# Twitter Airline Sentiment – Simple API Notebook

This notebook shows how to load the trained sentiment model from disk and
use it to predict the sentiment (negative / neutral / positive) of new tweets.
It assumes that `sentiment.example.ipynb` has already been run and that the
files `tfidf_vectorizer.joblib` and `logreg_sentiment_model.joblib` are present
in the same folder.


```python
import joblib

from sentiment_utils import clean_text, INV_LABEL_MAP

# Load artifacts saved by sentiment.example.ipynb
vectorizer = joblib.load("tfidf_vectorizer.joblib")
model = joblib.load("logreg_sentiment_model.joblib")

"Artifacts loaded."
```

```python
def predict_sentiment(text: str) -> str:
    """
    Predict sentiment for a single tweet string.

    Returns one of: 'negative', 'neutral', 'positive'.
    """
    clean = clean_text(text)
    vec = vectorizer.transform([clean])
    pred = model.predict(vec)[0]
    return INV_LABEL_MAP[pred]
```

## Example usage

The cell below shows how a client could call `predict_sentiment(text)`
with any tweet string and receive one of the three labels. In a real
application this function would sit behind an API endpoint or a small
web app.


```python
examples = [
    "I will never fly with this airline again.",
    "Pretty average flight, nothing special.",
    "Amazing crew and very comfortable seats!",
]

for t in examples:
    print(f"{predict_sentiment(t):>8}  |  {t}")
```
