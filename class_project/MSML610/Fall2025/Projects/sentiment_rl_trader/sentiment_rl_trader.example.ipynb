## Interpretation of Results

- **Final Return:** Indicates how much the capital grew
- **Win Rate:** Percentage of profitable trades
- **Sharpe Ratio:** Measures risk-adjusted return
- **Max Drawdown:** Largest peak-to-trough loss
- **Equity Curve:** Shows how capital evolved over time

These metrics help assess the quality of the reinforcement learning trading strategy.
