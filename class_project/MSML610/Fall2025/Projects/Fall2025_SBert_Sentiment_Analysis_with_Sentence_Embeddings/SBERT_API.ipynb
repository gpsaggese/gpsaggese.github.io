### Imports

```python
import pathlib, sys

# Make src/ importable
PROJECT_ROOT = pathlib.Path(".").resolve()
sys.path.append(str(PROJECT_ROOT / "src"))

from SBERT_utils import load_config, load_clean_data, load_embeddings
```

### Load config

```python
cfg = load_config("config.yaml")
df = load_clean_data(cfg)
print("Cleaned dataset shape:", df.shape)
df.head()
```

### Label distribution (simple sanity check)

```python
label_col = cfg["preprocessing"]["label_column"]
df[label_col].value_counts().sort_index()
```

### Load embeddings

```python
embeddings = load_embeddings(cfg)

print("Embeddings shape:", embeddings.shape)
# Quick consistency check
assert len(df) == embeddings.shape[0], "Rows in data and embeddings must match"
```

### Tiny API-style example

```python
i = 0

text_col = cfg["preprocessing"]["text_column"]
label_col = cfg["preprocessing"]["label_column"]

print("Sentence:", df.iloc[i][text_col])
print("Label:", df.iloc[i][label_col])
print("Embedding (first 10 dims):", embeddings[i][:10])
```
