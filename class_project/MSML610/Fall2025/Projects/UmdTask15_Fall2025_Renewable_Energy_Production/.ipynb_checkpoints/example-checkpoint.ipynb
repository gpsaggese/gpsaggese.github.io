```python
import pandas as pd
df = pd.read_csv("data/processed/train.csv", parse_dates=["timestamp"])
df.head()
```

```python
import matplotlib.pyplot as plt
ax = df.groupby("hour")["energy_mwh"].mean().plot()
ax.set_title("Average Solar Energy by Hour")
ax.set_xlabel("Hour of day"); ax.set_ylabel("MWh");
```

```python
day = df[df["hour"].between(6, 18)]
ax = day.plot(kind="scatter", x="cloud_cover", y="energy_mwh", alpha=0.25)
ax.set_title("Energy vs Cloud Cover (Daylight Hours)")
```

```python
import mlflow
import matplotlib.pyplot as plt
import pandas as pd

# point MLflow to the same local store we used in training
mlflow.set_tracking_uri("file:./mlruns")
mlflow.set_experiment("Solar_Forecasting")

# re-create the two figures you made above (so we can log them)
df = pd.read_csv("data/processed/train.csv", parse_dates=["timestamp"])

# Fig 1: average by hour
fig1 = plt.figure()
df.groupby("hour")["energy_mwh"].mean().plot()
plt.title("Average Solar Energy by Hour"); plt.xlabel("Hour"); plt.ylabel("MWh")

# Fig 2: cloud cover vs energy (daylight only)
day = df[df["hour"].between(6, 18)]
fig2 = plt.figure()
day.plot(kind="scatter", x="cloud_cover", y="energy_mwh", alpha=0.25)
plt.title("Energy vs Cloud Cover (Daylight Hours)")

with mlflow.start_run(run_name="notebook_eda"):
    mlflow.log_artifact("configs/baseline.yaml")
    mlflow.log_figure(fig1, "figures/avg_by_hour.png")
    mlflow.log_figure(fig2, "figures/energy_vs_cloud_daylight.png")
    mlflow.set_tag("stage", "EDA")
```
