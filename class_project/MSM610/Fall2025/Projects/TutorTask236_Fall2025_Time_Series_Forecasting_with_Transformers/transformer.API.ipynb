# Transformer Time-Series Forecasting — API Usage

This notebook demonstrates usage of the project’s native API and wrapper layer
to train, evaluate, and forecast stock prices using a Transformer model.


```python
# Imports
from transformer_utils import (
    load_data,
    train_and_evaluate,
    predict,
    compute_metrics,
)


```

```python
# Config
config = {
    "seq_len": 60,
    "horizon": 5,
    "epochs": 20
}
```

```python
# Load data
X_train, y_train, X_val, y_val, X_test, y_test = load_data(
    "data/AMZN.csv",
    config
)

```

```python
import os

config["ckpt_dir"] = os.path.abspath("./checkpoints")
```

```python
# Train
model, params = train_and_evaluate(X_train, y_train, X_val, y_val, config)
```

```python
# Predict
y_pred = predict(model, params, X_test)
```

```python
# Metrics
print(compute_metrics(y_test, y_pred))
```
