```python
# Cell 1: Import
from semantic_search_utils import SemanticSearchEngine, load_wikipedia_data
```

```python
# Cell 2: Load data
texts = load_wikipedia_data("data_sample.parquet", max_articles=1000)
print(f"Loaded {len(texts)} articles")
```

```python
engine = SemanticSearchEngine()
engine.load_model()
```

```python
engine.index_documents(texts)
```

```python
results = engine.search("famous tower in Paris", top_k=3)
for r in results:
    print(f"Score: {r.score:.3f} | {r.text[:100]}...")
```

```python
results = engine.search("artificial intelligence", top_k=3)
for r in results:
    print(f"Score: {r.score:.3f} | {r.text[:100]}...")
```
